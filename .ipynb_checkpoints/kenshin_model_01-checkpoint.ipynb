### 基本ライブラリのインポート

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### データのインポートとデータの整形

```python
df = pd.read_csv("signate_data/train.csv")
df.info()
```

```python
# データ整形関数
def mod_data(data):
    data = data.drop("id", axis=1)
    data["AG_ratio"] = data["AG_ratio"].fillna(data.Alb / (data.TP - data.Alb))
    data["Gender"] = data["Gender"].map({"Male": 0, "Female": 1})
    print(data.info())
    return data
```

```python
df = mod_data(df)
```

### データ確認

```python
# diseaseごとの説明変数の分布
plt.figure(figsize=(10, 10))
for ncol, colname in enumerate(df.columns):
    plt.subplot(4, 3, ncol+1)
    sns.histplot(df.query("disease==0")[colname], kde=True, color="blue")
    sns.histplot(df.query("disease==1")[colname], kde=True, color="orange")
    plt.legend(labels=["non", "diseased"], loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# 相関マップ
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), vmin=-1.0, vmax=1.0, annot=True, cmap='coolwarm', linewidths=0.1)
plt.show()
```

### モデルの作成

```python
X = df.drop("disease", axis=1)
y = df["disease"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# ロジスティックモデル
from sklearn.linear_model import LogisticRegression as LogiR
lr = LogiR(max_iter=1000)
lr.fit(X_train, y_train)
result = lr.predict(X_test)
```
